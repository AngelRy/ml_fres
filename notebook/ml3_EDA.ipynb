### 2.2 Dataset information

- studentID - integer 
- age       - int
- Ethnicity - int - Their ethnicity, coded as follows: - int - 0:Caucasian, 1:African American, 2:Asian, 3: Other
- gender    - boolean - 0 represents Male and 1 represents Female
- ParentalEducation: The ed level of the parents, coded as follows:
    0: None
    1: High School
    2: Some College
    3: Bachelor's
    4: Higher
- StudyTimeWeekly: Weekly study time in hours, ranging from 0 to 20
- Absences: Number of absences during the school year, ranging from 0 to 30.
- Tutoring: Tutoring status, where 0 indicates No and 1 indicates Yes.
- ParentalSupport: The level of parental support, coded as follows:
    0:None, 1:Low, 2:Moderate, 3:High, 4:VeryHigh

- Extracurricular: Participation in extracurricular activities, where 0 indicates No and 1 indicates Yes.
- Sports: Participation in sports, where 0 indicates No and 1 indicates Yes.
- Music: Participation in music activities, where 0 indicates No and 1 indicates Yes.
- Volunteering: Participation in volunteering, where 0 indicates No and 1 indicates Yes.

- GPA: Grade Point Average on a scale from 2.0 to 4.0, influenced by study habits, 
    parental involvement, and extracurricular activities.

- GradeClass: Classification of students' grades based on GPA:
        0: 'A' (GPA >= 3.5)
        1: 'B' (3.0 <= GPA < 3.5)
        2: 'C' (2.5 <= GPA < 3.0)
        3: 'D' (2.0 <= GPA < 2.5)
        4: 'F' (GPA < 2.0)

We will attempt, based on the given attribute values for every student (studentID and GradeClass excluded), to estimate their gpa as close as possible.

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 7))
plt.subplot(121)
sns.histplot(data=df,x='GPA',bins=19,kde=True,color='g')
plt.subplot(122)
sns.histplot(data=df,x='GPA',kde=True,hue='Gender')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2)
plt.subplot(121)
sns.boxplot(data=df, x="ParentalSupport", y="GPA")
plt.subplot(122)
sns.boxplot(data=df, x="ParentalSupport", hue="Gender", y="GPA")
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20,6))
plt.subplot(131)
sns.boxplot(data=df, x="ParentalSupport", y="GPA")
plt.subplot(132)
sns.boxplot(data=df, x="ParentalSupport", hue="Tutoring", y="GPA")
plt.subplot(133)
sns.boxplot(data=df, x="ParentalEducation", y="GPA")
plt.show()	
```

I the triple figure above, we see that the level of parental support has positive impact on the GPA, especially among the girls. The level of parental education, has no impact whatsoever - a negative one if any, when it comes to the group with highest parental education

```python
dfmv = df[['StudyTimeWeekly', 'Absences', 'GPA', 'Gender']]
sns.pairplot(dfmv, hue='Gender')
plt.show()
```

There's clear negative corellation between absences and GPA. Weekly study time has some impact in the extremes, meaning that no highest GPA for those that never study and no flunking for those that study maximum hours.

#### BIVARIATE ANALYSIS ( Does Tutoring, Extracurricular, Sports, Music, Volunteering have any impact on student's GPA? )

```python
plt.figure(figsize=(12,6))
plt.subplot(2,2,1)
sns.barplot (x=df['Tutoring'], y=df['math_score'], hue=df['test_preparation_course'])
plt.subplot(2,2,2)
sns.barplot (x=df['lunch'], y=df['reading_score'], hue=df['test_preparation_course'])
plt.subplot(2,2,3)
sns.barplot (x=df['lunch'], y=df['writing_score'], hue=df['test_preparation_course'])
```
